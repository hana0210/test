```python
# 레퍼런스 reference 

a = [1, 2, 3]
b = a # a 를 b에 밀어 넣는 식
b
b[0] = 10 # 1을 10으로 바꿈
b
a # 여기서 a까지 수정된 걸 볼 수 있다.
a[2] = 30
a
b # a를 바꿔도 b도 같이 바뀐걸 볼 수 있다.

# 이것을 a가 b를, b가 a를 "참조 reference"한다 라고 한다.
```

```python
a = [1, 2, 3,]
b = a.copy()
b[0] = 10

print(b)
print(a)

# 참조를 안하고 싶으면 이렇게 하면 된다.
```

```python
# except 를 더 자세히 작성해보면..

try:
    for i in range(4):
        print(li[i])
except IndexError: # except 뒤에 에러 이름을 써주면 된다.
    print("인덱스 에러가 생겼습니다")
```

```python
# except 를 더 자세히 작성해보면.. 교재 387 p
# 아래처럼 except를 여러개 써도 됨

try:
    for i in range(4):
        print(li[i])
except IndexError as e: # except 뒤에 에러 이름을 써주면 된다.
    # print("인덱스 에러가 생겼습니다")
    print(e)
except IndexError as e: # except 뒤에 에러 이름을 써주면 된다.
    # print("인덱스 에러가 생겼습니다")
    print(e)
except IndexError as e: # except 뒤에 에러 이름을 써주면 된다.
    # print("인덱스 에러가 생겼습니다")
    print(e)
except:
    print("뭔지 모르는 오류")
```

```python
def divide(a, b):
    return a/b

divide(10, 20)
divide(10, 10)
divide(10, 0)
```

```python
# 에러 처리를 해주세요.

try:
    divide(10, 0)
except:
    print("0으로 나눌 수 없습니다")
```

```python
# 에러 종류를 명시해서 처리하기

try:
    divide(10, 0)
except ZeroDivisionError as e:
    print(e)
```

```python
try:
    divide(10, 0)
except ZeroDivisionError as e:
    print(e)
except IndexError as e: # except 뒤에 에러 이름을 써주면 된다.
    # print("인덱스 에러가 생겼습니다")
    print(e)
finally:
    print("끝~")
```

```python
try:
    divide(10, 0)
except (IndexError, ZeroDivisionError) as e:
    print(e)
finally:
    print("끝~")
```
