````python
class Calculator:

    def __init__(self):
        print("객체를 만듭니다")

    def add(self, a, b):
        return a + b
        
    def subtract(self, a, b):
        return a - b
        
    def multiply(self, a, b):
        return a * b
        
    def divide(self, a, b):
        return a / b

cal = Calculator()
cal.add(1, 2)
cal.subtract(1, 2)
cal.multiply(1, 2)
cal.divide(1, 2)
````

````python
# 두 숫자는 __init__() 을 사용하여 입력받고,
# cal.add() 만 실행시켜서 합을 출력하도록.

class Calculator:

    def __init__(self, first, second):
        self.first = first
        self.second = second

    def add(self):
        return self.first + self.second

    def subtract(self):
        return self.first - self.second
        
    def multiply(self):
        return self.first * self.second
        
    def divide(self):
        return self.first / self.second
        
cal = Calculator(1, 2)
cal.add()
````

````python
cal.subtract()
````

````python
cal.multiply()
````

````python
cal.divide()
````
