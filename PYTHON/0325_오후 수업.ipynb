```python
from bs4 import BeautifulSoup
```

```python
# 1. 원하는 페이지에서 html 코드를 가져온다.

url = "https://www.pressian.com/pages/news-world-list?page=1"
html = urlopen(url).read()
html
```

```python
type(html)
```

```python
# 2. 다루기 쉽게 만들기

# Beautiful은 class 다.
soup = BeautifulSoup(html, 'html.parser') # parsing은 보기쉽게 만드는 거
soup
```

```python
type(soup)
```

```python
# 3. 뉴스의 타이틀들만 가져오기 - 제목 부분을 콕 찍는다 (class 사용) 

soup = BeautifulSoup(html, 'html.parser')
 # parsing은 보기쉽게 만드는 거
titles = soup.find_all(class_ = "title")
 # 클래스 뒤에는 언더스코어를 붙여줘야한다.

titles
# 리스트로 표기 됨
```

```python
# 특정 부분 부터 보고 싶을 때 

soup = BeautifulSoup(html, 'html.parser')
 # parsing은 보기쉽게 만드는 거
titles = soup.find_all(class_ = "title")[11:]
 # 클래스 뒤에는 언더스코어를 붙여줘야한다.
 # 11 번째 부터 끝까지 (인덱싱 넘버로 표기)

titles
```

```python
soup = BeautifulSoup(html, 'html.parser')
titles = soup.find_all(class_ = "title")[11:]

for title in titles:
    print(title.text)
    print()
```

```python
# 저장하기

for title in titles:
    with open("제목.txt", "a") as f:
        f.write(title.text)
        f.write("\n")
```

```python
# 여러 페이지를 보고 싶을 때

from urllib.request import urlopen
from bs4 import BeautifulSoup
import time

# 1. 원하는 페이지에서 html 코드를 가져온다
for i in range(1, 4):
    url1 = "https://www.pressian.com/pages/news-world-list?page="
    url2 = str(i)
    
    url = url1 + url2
    print(f"{url2}페이지를 가지고 옵니다")
    html = urlopen(url).read()
    
    # 2. 다루기 쉽게 만든다
    soup = BeautifulSoup(html, 'html.parser')
    
    # 3. 제목 부분을 콕 찍는다
    titles = soup.find_all(class_ = "title")[11:]
    
    # 4. 저장한다.
    for title in titles:
        with open("제목만.txt", "a", encoding = "utf-8") as f:
            f.write(title.text)
            print("\n")
            
    # 3. 본문 부분을 콕 찍는다
    bodies = soup.find_all(class_ = "body")
    
    # 4. 저장한다.
    for body in bodies:
        with open("내용만.txt", "a", encoding = "utf-8") as f:
            f.write(body.text)
            print("\n")

    time.sleep(1)
```
