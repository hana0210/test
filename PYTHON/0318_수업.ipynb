````python
def print_n_times(n, message):
    for i in range(n):
        print(message)

print_n_times(3, "안녕")
````

````python
def print_n_times(message, n): # 이 라인에서 message 하고 n 위치가 변경되어도 됨.
    for i in range(n):
        print(message)

print_n_times("안녕", 3) # 이 라인에서 message 하고 n 위치가 변경되어도 됨.

# print_n_times(message = "안녕", n = 3) > 이렇게도 씀.
````

````python
# 두수를 argument 로 입력받아서 합을 구하는 함수

def total(start, stop):
    total = 0
    for i in range(start, stop+1):
        total += i
    return total

total(1234, 12345)
````

````python
# 디폴트(기본값) 주기
# 디폴트 값이 있는 것은 뒤에 적어주게 되어 있다.

def total(start,stop, step=1):
    total = 0
    for i in range(start, stop+1, step):
        total += i
    return total     
total(2, 10)

# 값 두개만 줘도 작동함.
````

````python
# 0 ~ 입력받은 수까지 어떤수의 배수를 제외한 모든 수의 합

a = 0
total = 0
while a < 10:
    a = a + 1
    if a % 3 == 0:
        continue
    total = total + a

total
````

````python
# 위에 거 def으로 해보기

def sum1(a, b):
    a = 0
    total = 0
    while a < 10:
        a = a + 1
        if a % b == 0:
            continue
        total = total + a
    return total

sum1(10, 3)
````

````python
# 인수의 개수가 정해져 있지 않을 때
# * 를 해준다 > * 리스트(여러개)라는 뜻이다

def sum2(*a):
    for i in a:
        print(i)

sum2(1, 2, 3, 4)
````

````python
def sum2(*a):
    total = 0
    for i in a:
        total += i
    return total

sum2(1, 2, 3, 4, 5, 6, 7)
````

````python
# 1,2 는 두고 
# args = arguments
# 

def test(name, age, *args):
    total = 0
    for i in args:
        total += i
    return f" {name}님의 나이는 {age}이고 총점은 {total}입니다."

test("길동", 23, 25, 27, 30, 100, 35)

# 교재 278 ~ 279 
````

````python
# 입력받은 수를 2배해서 내놓는 함수 : twice

def twice(X):
    result = X*2
    return result

twice(10)
````

````python
# 두 수를 입력받아서 두수의 합을 출력하는 함수

def plus(a, b):
    return a + b

plus(1, 2)
````

````python
# 하나를 입력받아서 3보다 큰지 작은지를 참 거짓으로 판단해주는 함수

def check(n):
    if n > 3:
        return True
    else:
        return False

check(1)
````

````python
def check(n): # 하나의 수를 입력받아서
    return n > 3 # 3보다 큰지 ?

check(5)
````
