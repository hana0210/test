```python
# 0 부터 입력받은 수 까지의 합

number = int(input("숫자를 입력하세요"))
total = 0

for i in range(number+1):
    total += i
print(total)
```

```python
# 13 ~ 123까지 홀수의 합과 개수는 ?

total = 0
count = 0

for i in range(12, 123+1):
    if i % 2 == 1:
        total += i
        count += 1
print(total)
print(count)
```

```python
start = int(input("시작하는 숫자: "))
end = int(input("끝나는 숫자: "))

total = 0
count = 0

for i in range(start, end+1):
    if i % 2 == 1:
        total += i
        count += 1
print(f"{start} 부터 {end}까지의 홀수의 합은 {total}입니다")
print(count)
```
