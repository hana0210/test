````python
# Q2.  각 프로들의 스코어 합계는?

class GolfScore:
    def __init__(self, name, scores):
        self.name = name
        self.scores = scores
        print(f"{self.name}님의 현재 스코어: {self.scores}")

    def total(self):
        score_sum = sum(self.scores)
        
        if score_sum == 0:
            return "Even"
        elif score_sum > 0:
            return f"+{score_sum}"
        else:
            return f"{score_sum}"

p1 = GolfScore("윤프로", [-1, 0])
p2 = GolfScore("이프로", [0, 0])
p3 = GolfScore("장프로", [-2, -1])

players = [p1, p2, p3]
print("\n--- 경기 결과 ---")
for p in players: print(f"{p.name} 최종 결과: {p.total()}")
````

````python
class GolfScore:
    def __init__(self, name, scores):
        self.name = name
        
        # 데이터가 리스트가 아니면 리스트로 변환 (isinstance 활용)
        self.scores = scores if isinstance(scores, list) else [scores]
        print(f"⛳ {self.name}님 등록 완료: {self.scores}")

    def get_total_int(self):
        """계산을 위해 정수형 합계를 반환"""
        return sum(self.scores)

    def total_display(self):
        """출력용 포맷팅 (Even, +1, -2 등)"""
        s = self.get_total_int()
        if s == 0: return "Even"
        return f"{s:+d}"

# 1. 선수 데이터 등록
players = [
    GolfScore("윤프로", [-1, 0, -1]), # -2
    GolfScore("이프로", [0, 1, 0]),    # +1
    GolfScore("장프로", [-2, -1, 0])  # -3
]

print("-" * 30)

# 2. 모든 선수 결과 출력 (한 줄 for문)
for p in players: print(f"{p.name} 최종 결과: {p.total_display()}")

print("-" * 30)

# 3. 우승자 찾기 (합계 점수가 가장 낮은 사람)
# min 함수와 key를 사용하여 합계가 최소인 객체를 찾습니다.
winner = min(players, key=lambda p: p.get_total_int())

print(f"🏆 오늘의 우승자는 {winner.name}님입니다! (스코어: {winner.total_display()})")
````

- 개수(Count): sum 1 (조건이 맞을 때마다 1씩 상수를 더함)
- 합계(Sum): sum i (변수 $i$의 값 자체를 누적해서 더함)

````python
# 12 부터 30까지 모든 수의 합과
# 평균을 출력해봅시다.

total = 0 
count = 0

for i in range(12, 31):
    total = total + i
    count = count + 1
print(f"합:{total}, 평균:{total/count}")
````

````python
# 다른방법 > 함수로

def calc(a, b):
    total = 0
    count = 0
    for i in range (a, b + 1):
        total = total + i
        count += 1
    print(f"합:{total}, 평균:{total/count}")

calc(12, 30)
````

````python
# 12 부터 30 까지 모든 홀수의 합과 평균을 출력해봅시다.

total = 0
count = 0

for i in range(12, 31):
    if i % 2 == 1: # 2로 나눈 나머지가 홀수라면..
        total = total + i
        count += 1

print(f"합:{total}, 평균:{total / count}")
````

````python
# 홀수만 출력해보자

di = {"A" : 1, "B" : 2, "C" : 3, "D" : 4}

for i, j in di.items():
    if j % 2 == 1:
        print(j)
````

````python
# dictionary 에서 value 가 홀수인 Key 만 출력

for i, j in di.items():
    if j % 2 == 1:
        print(i)
````

````python
# dictionary 에서 value 가 홀수인 Key 를 'odd' 안에 리스트 형태로 모아보자

odd = []

for key, value in di.items():
    if value % 2 == 1:
        odd.append(key)

odd
````

````python
# dictionary 에서 value 가 홀수인 Key 를 홀수 짝수 구분하여 출력해보자

odd = []
even = []

for key, value in di.items():
    if value % 2 == 1:
        odd.append(key)
else:
    even.append(key)

print(odd)
print(even)
````
