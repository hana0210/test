```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import time 
# 1. 원하는 페이지에서 html 코드를 가져온다

for i in range(1, 20):    
    url1 = "https://www.pressian.com/pages/news-world-list?page="
    url2 = str(i)
    url = url1 + url2
    print(f"{url2}페이지를 가지고 옵니다")    
    html = urlopen(url).read()
    
    # 2. 다루기 쉽게 만든다
    soup = BeautifulSoup(html, 'html.parser')
    
    # 3. 제목 부분을 콕 찍는다
    titles = soup.find_all(class_ = "title")[11:]
    
    # 4. 저장한다.
    for title in titles:
        with open("제목만.txt", "a", encoding="utf-8") as f:
            f.write(title.text)
            f.write("\n")
       
        
    # 3. 본문 부분을 콕 찍는다
    bodies = soup.find_all(class_ = "body")
    
    # 4. 저장한다.
    for body in bodies:
        with open("내용만.txt", "a", encoding="utf-8") as f:
            f.write(body.text)
            
    time.sleep(1)
```
