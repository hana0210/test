### 상속 inheritance

````python
# 교재 496 ~500

class Calculator:
    
    def add(self, a, b):
        return a + b
        
    def subtract(self, a, b):
        return a - b
        
    def multiply(self, a, b):
        return a * b
        
    def divide(self, a, b):
        return a / b
        
````

````python
cal = Calculator() # 객체 만들기
````

````python
cal.add(1, 2) # cal. tab을 누르면 사용할 수 있는 함수가 뜬다.
````

````python
cal.divide(1, 2)
````

````python
# 거듭 제곱을 만들고 싶을 때..

class Child(Calculator): # 가지고 오고 싶은 클래스를 괄호안에 쓴다.
    def power(self, a, b):
        return a**b

ch1 =  Child()
# 이렇게 하면 설계도가 완성

print(ch1.power(2, 3)) # 작동하는지 확인하려면.. 
print(ch1.add(1,2)) # 여기서 만들지 않은 함수가 작동할까?

# 부모(calculator) 가 가지고 있는 함수는 다 사용 가능

print(dir(ch1)) # 사용가능한 함수를 확인해보려면
````

````python
# calculator (본점) 에서 만든 'divide' 가 마음에 들지 않아 Child (지점) 에서 수정 하기

# override, overriding < 메소드 재정의 // 동사보다는 현재진행형으로 많이 씀.
# 같은 이름의 함수가 있을 때는 자식(지점) 클래스의 함수가 실해오딘다.

class Child(Calculator): # 가지고 오고 싶은 클래스를 괄호안에 쓴다.
    def power(self, a, b):
        return a**b

    def divide(self, a, b):
        if b == 0:
            return "0으로 나눌 수 없습니다"
        return a/b

ch1 =  Child()
ch1.divide(1, 0)
````

````python
## 상속 2

class Student():
    school = "신림중학교" 
    
    def __init__(self, name, age):
        self.name = name        # 인스턴스 변수
        self.age = age          # local variable 로컬 변수와 같은

    def sing(self):
        print(f"{self.name} 이 노래를 불러요")

    def __str__(self):
        return "학생 객체 입니다."
````

````python
class Grade(Student): # Grade 는 Student 를 상속 받았다는 뜻
    
    def __init__(self, name, age, grade): # 2번 아래 있는거 포함해서 argument 를 줘야한다.
        super().__init__(name, age) # 1번 위에 있는 init 함수를 써라 ; super 가 부모를 지칭하는 단어
        self.grade = grade

    def show(self):
        return f"{self.name}는 {self.grade}학년 입니다"

s = Grade("철수", 15, 3)
s.show()
````
