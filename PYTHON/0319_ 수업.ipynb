````python
def hello(n):
    if n == 0:
        return 
    print("Hello")
    time.sleep(1)
    n = n - 1
    hello(n)

hello(5)
````

````python
# 5! = 5 factorial = 5 * 4 * 3 * 2 * 1
# 5! = 5 * 4! = 5 * 4 * 3!

def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n-1)

factorial(5)
# 5 * 4 * 3 * 2 * factorial(1)

# 교재 293 ~
````

````python
def fac(n):
    result = 1
    for i in range(1, n + 1):
        result = result * i
    return result

fac(5)
````

````python
# 올바른 시간 측정 방법
# 전체 실행 시간을 알고 싶다면, 시간 측정 로직을 함수 외부로 꺼내야 합니다.

import time

def fibo(n):
    if n == 1 or n == 2:
        return 1
    return fibo(n-1) + fibo(n-2)

# 1. 시작 시간 기록
start_time = time.time()

# 2. 함수 실행
result = fibo(35) # 20은 너무 빨라서 35 정도로 테스트해보세요!

# 3. 종료 시간 기록 및 차이 계산
end_time = time.time()

print(f"결과: {result}")
print(f"전체 소요 시간: {end_time - start_time:.4f}초")
````

````python
# 함수가 실제로 몇 번 호출되는지 직접 확인해보는 코드

def fibo(n):
    global count
    count += 1
    if n == 1 or n == 2:
        return 1
    return fibo(n-1) + fibo(n-2)
    
count = 0 

print(fibo(10))
print(count)

#10번째 숫자를 구하는데 함수가 무려 109번이나 호출
````

````python
# enumerate()

li = ['a', 'b', 'c', 'd']
list(enumerate(li))
````

````python
# 번호 매기고 싶을 때

for i, j in enumerate(li):
    print(f"번호: {i+1}, 이름: {j}")
````

````python
# 번호 매기고 싶을 때 (스타팅 번호를 지정 해주는 것도 됨)

for i, j in enumerate(li, start = 1):
    print(f"번호: {i}, 이름: {j}")
````
